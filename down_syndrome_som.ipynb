```python
from sade.datasets.loaders import get_image_files_list
extract_id = lambda f: f.split("/")[-1].replace('.nii.gz', '')

dataset_dir = "/DATA/Users/amahmood/braintyp/processed_v2/"
splits_dir = "/ASD/ahsan_projects//braintypicality/dataset/dataset_split_builder/"

abcd_train_files = get_image_files_list("abcd-train", dataset_dir, splits_dir)
abcd_val_files = get_image_files_list("abcd-val", dataset_dir, splits_dir)
abcd_test_files = get_image_files_list("abcd-test", dataset_dir, splits_dir)

ibis_inlier_files = get_image_files_list("ibis-inlier", dataset_dir, splits_dir)
ibis_ds_files = get_image_files_list("ibis-ds-sa", dataset_dir, splits_dir)
ibis_hr_files = get_image_files_list("ibis-hr-inlier", dataset_dir, splits_dir)
ibis_atyp_files = get_image_files_list("ibis-atypical", dataset_dir, splits_dir)
ibis_asd_files = get_image_files_list("ibis-asd", dataset_dir, splits_dir)
```

```python
dataset_id_to_name = {-1: "ABCD", 0:"LR-Typical", 1:"HR-Typical", 2:"Atypical", 3:"Down Syndrome", 4:"ASD"}
dataset_name_to_id = {v:k for k,v in dataset_id_to_name.items()}
dataset_list = [abcd_data, ibis_typical, ibis_hr_typical, ibis_atypical, ibis_ds, ibis_asd]

X_data = np.concatenate(dataset_list, axis=0)
y_labels = np.concatenate([[i-1]*len(d) for i,d in enumerate(dataset_list)])

assert len(y_labels)==len(X_data)
# X_data = (X_data - np.mean(abcd_data, axis=0)) / np.std(abcd_data, axis=0)
inlier_data = X_data[y_labels <= 0]
test_data = X_data[y_labels > 0]
test_labels = y_labels[y_labels > 0]


mu, std = np.mean(inlier_data, axis=0), np.std(inlier_data, axis=0)
inlier_data = (inlier_data - mu) / std
test_data = (test_data  - mu) / std

dssamples = test_data[test_labels == dataset_name_to_id["Down Syndrome"]]
X_data.shape, len(dssamples), test_data.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5).fit(inlier_data)
print(pca.explained_variance_ratio_, pca.explained_variance_ratio_[0]/pca.explained_variance_ratio_[1])

num_neurons = 5*np.sqrt(inlier_data.shape[0])
grid_size = int(np.ceil(np.sqrt(num_neurons)))

ratio = np.floor(pca.explained_variance_ratio_[0]/pca.explained_variance_ratio_[1])
h = np.ceil(np.sqrt(num_neurons / ratio))
w = h*ratio
print(f"Suggested minimum number of neurons: {int(num_neurons)}")
print(f"Suggested height and width: {int(h)}x{int(w)}")
```

```python
%%time
som_height, som_width = 13, 13
net = sps.SOMNet(som_height, som_width, inlier_data, topology='hexagonal', 
                init='PCA', metric='euclidean',
                neighborhood_fun='gaussian', PBC=True,
                random_seed=42, GPU=True, CUML=False,
                output_path="/tmp/")

net.train(train_algo='batch', start_learning_rate=0.1, epochs=10000, batch_size=128,
           early_stop='mapdiff', early_stop_tolerance=1e-5, early_stop_patience=50
)
net.plot_convergence(figsize=(11, 5), logy=True)
net.get_nodes_difference()
net.plot_map_by_difference(show=True, print_out=False)
```

```python
# %%time
bmus_arr = net.find_bmu_ix(cupy.asarray(test_data)).get()
bmu_dists = [np.linalg.norm(net.nodes_list[bmuidx].weights - x) for bmuidx,x in zip(bmus_arr, test_data)]
```

```python
# %%time
bmus = pd.DataFrame([b.item() for b in bmus_arr], columns=["BMU"])
bmus['dataset'] = [dataset_id_to_name[i] for i in test_labels]
bmus['bmu-distance'] = bmu_dists
bmus.head(3)
```

```python
bmus.shape
```

```python
dataset_counts = bmus['dataset'].value_counts()
dataset_counts
```

```python
maxbmus = bmus.groupby(['BMU']).aggregate(dataset = ('dataset', lambda c: pd.Series.mode(c)[0]),
                                          best_match_idx=('bmu-distance', pd.Series.idxmin))

vcounts = bmus[['BMU', 'dataset']].groupby('BMU').value_counts()
total  =  bmus[['BMU', 'dataset']].groupby('BMU').count()

for idx in maxbmus.index:
    maxbmus.loc[idx, 'count'] = vcounts[(idx, maxbmus.loc[idx, 'dataset'])]
    maxbmus.loc[idx, 'total'] = total.loc[idx, 'dataset']
    
    node = net.nodes_list[idx]
    maxbmus.loc[idx, ['x', 'y']] = node.pos.get()
    maxbmus.loc[idx, 'neighbourhood-distance'] = node.difference

maxbmus.head(3)
```

```python
bmucounts = bmus[['BMU', 'dataset']].groupby(['BMU']).value_counts().reset_index(level=-1, name='count')
bmucounts
```

```python
for idx in bmucounts.index:

    node = net.nodes_list[idx]
    bmucounts.loc[idx, ['x', 'y']] = node.pos.get()
    bmucounts.loc[idx, 'neighbourhood-distance'] = node.difference

bmucounts.tail(3)
```

```python
significant_bmus = []
for name, g in bmucounts.groupby('dataset'):
    cohort_count = dataset_counts[name]
    min_cohort_count = int(0.1*cohort_count)
    print(name, min_cohort_count)
    significant_bmus.append(g[g['count'] >= int(0.1*cohort_count)])

significant_bmus = pd.concat(significant_bmus).sort_values('count', ascending=False)
significant_bmus
```

```python
HEXSIZE = .99 / np.sqrt(3)
marker_dict = OrderedDict({"Down Syndrome":"triangle", "ASD":"star", "HR-Typical":"circle", "Atypical":"diamond"})
DATASETS = list(marker_dict.keys())
MARKERS = list(marker_dict.values())

x = significant_bmus.x
y = significant_bmus.y
dataset = significant_bmus['dataset']
count = significant_bmus['count']

q,r = cartesian_to_axial(x,y, size=HEXSIZE, orientation="pointytop", aspect_scale=1)
sx,sy = axial_to_cartesian(q,r,  size=1, orientation="pointytop")

hexsource = ColumnDataSource(dict(q=q,r=r, dataset=dataset,  scale=scale(count, 0.1, 1.1)))
scattersource = ColumnDataSource(dict(x=sx,y=sy, dataset=dataset, scale=count+10))
group_filter = BooleanFilter([c != "LR-Typical" for c in dataset])
view = CDSView(filter=group_filter)
```

```python
plot = figure(width=450, height=400)
renderer = plot.hex_tile(
    source=hexsource, view=view,
    color=factor_cmap('dataset', 'Category10_7', DATASETS),
    scale='scale',
    legend_group="dataset"
)

plot.legend.location = "top_left"
plot.legend.orientation = "horizontal"
# plot.legend.title = "Dataset"
show(plot)
```

### Computing background hex tiles

```python
bmu_density = net.find_bmu_ix(cupy.asarray(inlier_data)).get()
bmu_density = Counter(bmu_density)
plt.hist(bmu_density.values());
plt.xlabel("Number of samples in Node")
```

```python
x,y,distance, sample_count, density = [],[],[],[], []
for idx, n in enumerate(net.nodes_list):
    x.append(n.pos[0].get())
    y.append(n.pos[1].get())
    distance.append(n.difference)
    c = bmu_density.get(idx, 0)
    sample_count.append(c)
    density.append(100 * c / len(inlier_data))

x,y,distance = np.asarray(x),np.asarray(y),np.asarray(distance)
q,r = cartesian_to_axial(x,y, size=HEXSIZE, orientation="pointytop")
binsource = ColumnDataSource(dict(q=q,r=r, distance=distance, density=density))
```

```python
#from color brewer
color_dict = OrderedDict({"Down Syndrome":'#e41a1c', #red
                          "HR-Typical":"#7fc97f", #green
                          "ASD":"#fdc076", #orange
                          "Atypical":"#a6cee3", #blue
                         })
```

```python
bmusdf = bmus.copy().set_index('BMU')

for idx in bmusdf.index:
    node = net.nodes_list[idx]
    bmusdf.loc[idx, ['x', 'y']] = node.pos.get()
bmusdf
```

```python
SCALE_FACTOR = 1
background_fill_color="#fafafa"
header_text_size = "18px"
width, height = 800, 700
# width, height = 650, 250 + 50

_x,_y = bmusdf.x, bmusdf.y
_q,_r = cartesian_to_axial(_x,_y, size=HEXSIZE, orientation="pointytop", aspect_scale=1)
_sx,_sy = axial_to_cartesian(_q,_r,  size=1, orientation="pointytop")

scattersource = ColumnDataSource(dict(x= _sx, y=_sy, dataset=bmusdf['dataset'],
                                color=[color_dict[d] for d in bmusdf['dataset']]
))
group_filter = BooleanFilter([c != "LR-Typical" for c in bmusdf['dataset']])
view = CDSView(filter=group_filter)

plot = figure(width=int(width * SCALE_FACTOR), height=int(height * SCALE_FACTOR), background_fill_color=background_fill_color)
r = plot.hex_tile(
    source=binsource, color="darkgray",
    fill_color=linear_cmap("distance", 'Greys256', 0.9*min(distance), max(distance)),
    alpha=0.8
)
color_bar = r.construct_color_bar(padding=1, title="Mean Neighbour Distance",
                                 title_text_font_size=f"{18*SCALE_FACTOR}px",
                                 major_label_text_font_size="0pt",
                                 major_tick_line_color=None
                                 )
plot.add_layout(color_bar, "right")

plot.scatter(source=scattersource, view=view,
             marker=factor_mark('dataset', MARKERS, DATASETS),
             color='color',
#              color=factor_cmap('dataset', 'Accent3', DATASETS),
             x=jitter('x', .9),
             y=jitter('y', .9),
             size=8*SCALE_FACTOR,
             legend_group="dataset",
             alpha=0.8
)

legend = Legend(
    items=plot.legend.items,
    location='top_right',
    orientation='horizontal',
    title='Dataset',
    title_text_font_size=f"{int(14*SCALE_FACTOR)}px",
    background_fill_color=background_fill_color,
    label_text_font_size=f"{int(13*SCALE_FACTOR)}px",
    glyph_height=int(32*SCALE_FACTOR),
    glyph_width=int(32*SCALE_FACTOR)
)

# Remove original legend
plot.legend.destroy()
plot.add_layout(legend, 'above')

plot.axis.visible = False
plot.grid.visible = False
distance_plot = plot
show(plot)
```
